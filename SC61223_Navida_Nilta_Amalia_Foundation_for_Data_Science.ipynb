5. Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.
Normal: 18.5-24.9
Overweight: 25.0-29.9
Obesity: 30.0-34.9
Underweight: < 18.5

```
print("--No.5--\n")
import pandas as pd

# Create a new column for BMI
df_12['BMI'] = df_12['Berat Badan (kg)'] / (df_12['Tinggi Badan (cm)'] / 100)**2

# Create a new column for Status BMI
df_12['Status BMI'] = ''

# Define category thresholds (numerical values)
underweight_threshold = float('< 18.5'.split()[1])  # Extract and convert numerical value
normal_thresholds = [float(x) for x in '18.5-24.9'.split('-')]  # Convert to a list of floats
overweight_thresholds = [float(x) for x in '25.0-29.9'.split('-')]  # Convert to a list of floats
obesity_thresholds = [float(x) for x in '30.0-34.9'.split('-')]  # Convert to a list of floats

# Implement If Else for determining Status BMI
for index, row in df_12.iterrows():
  if row['BMI'] < underweight_threshold:
    df_12.loc[index, 'Status BMI'] = 'Underweight'
  elif row['BMI'] >= normal_thresholds[0] and row['BMI'] < normal_thresholds[1]:  # Use list elements
    df_12.loc[index, 'Status BMI'] = 'Normal'
  elif row['BMI'] >= overweight_thresholds[0] and row['BMI'] < overweight_thresholds[1]:  # Use list elements
    df_12.loc[index, 'Status BMI'] = 'Overweight'
  else:
    df_12.loc[index, 'Status BMI'] = 'Obesitas'

df_12
```

2. Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.

Contoh:
Input
>> Money Capital : 100000
>> Investment Tenure (years) : 2
Output
>> Year 1, your total investment : 125000.0
>> Year 2, your total investment : 156250.0

```
print("--No.2--\n")
def hitung_return_investasi(modal, jangka_waktu):
    if not 0 < modal <= 1000000000:
        print("Error: Modal investasi harus dalam rentang 1 hingga 1.000.000.000")
        return
    if not 0 < jangka_waktu <= 25:
        print("Error: Jangka waktu investasi harus dalam rentang 1 hingga 25 tahun")
        return

    total_investasi = modal
    print("Tahun\tAkumulasi Investasi")
    for tahun in range(1, jangka_waktu + 1):
        total_investasi *= 1.25  # Menambahkan keuntungan tahunan
        print(f"{tahun}\t{total_investasi:.2f}")


modal = int(input("Masukkan modal investasi (1-1.000.000.000): "))
jangka_waktu = int(input("Masukkan jangka waktu investasi (1-25 tahun): "))

hitung_return_investasi(modal, jangka_waktu)
```
